```python
import pandas as pd
import unicodedata
import re
```

```python
# Cleaning the data
# Instead of the typical lemmatization and stemmatization, I think it would be wise to retain the morphological meanings 
# by keeping the endings. This matters especially for ancient Greek where everything is compound word.

with open('data/combined_text.txt', 'r', encoding='utf-8') as f:
    text = f.read()

print(type(text))

def clean_greek(text):
    # Normalize to NFC (Standard for Greek)
    text = unicodedata.normalize('NFC', text)
    # Lowercase (Ancient Greek has distinct case, but for embeddings, lowercase is usually safer)
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^\u0370-\u03FF\u1F00-\u1FFF\s]', '', text)
    # Optional: Strip accents if your dataset is small (comment out if you want to keep them)
    # text = "".join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    return text

# Example usage

cleaned_data = clean_greek(text)
```

```python
with open('data/cleaned_combined_text.txt', 'w',encoding='utf-8') as f:
    f.write(cleaned_data)
```
