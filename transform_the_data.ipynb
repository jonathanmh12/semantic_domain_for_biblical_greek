```python
import os
import re
from pathlib import Path

input_dir = Path('./data/')
output_dir = input_dir / 'cleaned'

output_dir.mkdir(parents=True, exist_ok=True)

pattern = r"^[1-9]?[A-Za-z]+\s\d+:\d+\s+"

for file_path in input_dir.glob('*.txt'):
    
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    cleaned_lines = [re.sub(pattern, "", line) for line in lines]

    new_file_path = output_dir / file_path.name

    with open(new_file_path, 'w', encoding='utf-8') as f:
        f.writelines(cleaned_lines)
        
    print(f"Processed: {file_path.name} -> {new_file_path}")

print("\nAll files cleaned and stored in /data/cleaned/")
```

```python
input_dir = Path('./data/cleaned_by_verse/')
output_dir = Path('./data/cleaned_by_book/')
output_dir.mkdir(parents=True, exist_ok=True)

for file_path in input_dir.glob('*.txt'):
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    single_line_text = " ".join(line.strip() for line in lines)
    
    new_file_path = output_dir / file_path.name

    with open(new_file_path, 'w', encoding='utf-8') as f:
        f.write(single_line_text)
    
    print(f"Compressed {file_path.name} into one line.")

print("Done! All books are now single-line files.")
```

```python
input_dir = Path('./data/cleaned_by_verse/')
output_file = Path('./data/combined_text.txt')

with open(output_file, 'w', encoding='utf-8') as master_f:
    
    for file_path in sorted(input_dir.glob('*.txt')):
        
        with open(file_path, 'r', encoding='utf-8') as f:
            lines = f.readlines()

        single_line_book = " ".join(line.strip() for line in lines)
        
        master_f.write(single_line_book + "\n")
        
        print(f"Added {file_path.name} to the master file.")

print(f"\nSuccess! Everything is combined in: {output_file}")
```
